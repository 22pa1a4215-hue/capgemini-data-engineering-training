```python
%sql
SELECT emp_name, salary,
ROW_NUMBER() 
  OVER (ORDER BY salary Desc) AS row_number
From employees3;
```

```python
%sql
SELECT emp_name, department,
ROW_NUMBER() 
  OVER (partition by department ORDER BY salary Desc) AS row_number
From employees3;
```

```python
%sql
SELECT emp_name, join_date,
row_number()
  OVER (ORDER BY join_date desc) AS row_number
From employees3
  
```

```python
%sql
SELECT emp_name, join_date,
row_number()
  OVER (ORDER BY join_date asc) AS row_number
From employees3
```

```python
%sql
SELECT emp_name, salary,
RANK()
  OVER (ORDER BY salary Desc) AS rank
From employees3
```

```python
%sql
SELECT emp_name, department,
RANK()
  OVER (partition by department ORDER BY salary Desc)
  AS rank
From employees3
```

```python
%sql
SELECT emp_name, join_date,
RANK()
 OVER(ORDER BY join_date DESC)
 AS rank
FROM employees3
```

```python
%sql
SELECT emp_name, department, salary,
RANK()
  OVER (partition by department ORDER BY salary Desc)
  AS rank
FROM employees3
```

```python
%sql
SELECT emp_name, salary,
DENSE_RANK()
  OVER (ORDER BY salary DESC)
AS RANK
FROM employees3
```

```python
%sql
SELECT emp_name, join_date,
DENSE_RANK()
  OVER (ORDER BY join_date asc)
AS RANK
FROM employees3
```

```python
%sql
SELECT emp_name, salary,
DENSE_RANK()
 OVER(ORDER BY salary asc)
 AS rank
FROM employees3
```

```python
%sql
SELECT department, join_date,
DENSE_RANK()
 OVER(partition by department ORDER BY join_date asc)
 AS rank
FROM employees3
```
